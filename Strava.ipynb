### **Extraindo colunas importantes para insights**

```python
colunas = ['Activity ID', 'Activity Date', 'Activity Name', 'Activity Type','Moving Time', 'Elapsed Time', 'Distance', 'Max Speed', 'Average Speed','Number of Runs']
```

```python
df = df[colunas].copy()
df.head()
```

```python
df['Activity date'] = pd.to_datetime(df['Activity Date']) #Transformando a data para melhor visualização
print(df.head())
```

```python
df['Moving Time Minutos'] = df['Moving Time'] / 60  #Transformando tempo de corrida de segundos para minutos
print(df.head())
```

```python
df['Average speed'] = 60 / (df['Average Speed'] * 3.6) #Transformando média da corrida de m/s para km/h, e depois km/h para km/min (Pace de corrida)
print(df.head())
```

```python
remover = ['Average Speed','Moving Time','Number of Runs','Activity Date','Elapsed Time'] #Removendo colunas antigas duplicadas, deixando apenas as que foram transformadas
df = df.drop(remover, axis=1)
```

```python
corridas_df = df[df['Activity Type'] == 'Run'] #Filtra para deixar a coluna com apenas atividades de corrida
numero_de_corridas = len(corridas_df)
print(numero_de_corridas)
```

```python
corridas_df = corridas_df.round(2)
corridas_df.head() #Verificando a situação dos dados
```

```python
df_pronto = 'activities_cleaned.csv' #Salvando o arquivo limpo e pronto para extrair insights visuais
corridas_df.to_csv(df_pronto, index = False)
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# total de dias que correu por mês
# a media de minutos corridos por semana
# pace
# distância corrida total
# distância corrida por semana

df_pronto = pd.read_csv('/content/activities_cleaned.csv')
```

```python
primeiro_dia = df_pronto['Activity date'].min() # Algumas curiosidades
ultimo_dia = df_pronto['Activity date'].max()
print(f"Primerio dia marcado - {primeiro_dia}")
print(f"Último dia marcado - {ultimo_dia}")
```

```python
numero_corridas = len(df_pronto)  # Total de corridas
print(f"Número de corridas totais - {numero_corridas}")
```

```python
distancia_total = df_pronto['Distance'].sum() # distância percorrida total
print(f"A distância total percorrida é {distancia_total} Km")
```

```python
tempo_total = df_pronto['Moving Time Minutos'].sum() # Tempo total(geral)
tempo_total = round(tempo_total,2)
print(f"Tempo total de corrida é {tempo_total} Minutos")
```

```python
pace_geral = tempo_total / distancia_total  # Pace
pace_geral = round(pace_geral,2)
print(f"A Pace Geral é {pace_geral} por km")
```

```python
corridas_df.set_index('Activity date', inplace=True) # Deixando Série Temporal
distancia_semana = corridas_df['Distance'].resample('W-SUN').sum().reset_index()  # distância percorrida por semana
distancia_semana.columns = ['Semana', 'Distância']
print(distancia_semana)
```

```python
import plotly.express as px

fig = px.bar(
    distancia_semana,
    x='Semana',
    y='Distância',
    labels={
        "Semana": "Semana (Início no Domigo)",
        "Distância": "Distância Total (km)"
    },
    title="Distância Percorrida por Semana",
    color_discrete_sequence=['#FC4C02'],
)
fig.update_layout(
    title_x=0.5,
    xaxis_title="Período de Treino",
    yaxis_title="Distância Total (km)",
    plot_bgcolor='#C0C0C0'
)

fig.update_xaxes(dtick="M1", tickformat="%d %b\n%Y")

fig.show()
```

```python
minutos_semana = corridas_df['Moving Time Minutos'].resample('W-SUN').sum().reset_index() # a media de minutos corridos por semana
minutos_semana.columns = ['Semana', 'Tempo_minutos']
print(minutos_semana)
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.line(
    minutos_semana,
    x='Semana',
    y='Tempo_minutos',
    title='Total de Minutos Corridos por Semana',
    labels={
        "Semana": "Data",
        "Tempo_minutos": "Minutos Corridos na Semana"},
    color_discrete_sequence=['#FC4C02']
    )

fig.update_xaxes(rangeslider_visible=True)

fig.update_layout(
    title_x=0.5,
    xaxis_title="Período de Treino",
    yaxis_title="Tempo Total (minutos)",
    plot_bgcolor='#C0C0C0')


fig.show()
```

```python
dias_por_mes = corridas_df.resample('ME').size() # total de dias que correu por mês
dias_por_mes = dias_por_mes.reset_index()
dias_por_mes.columns = ['Mês', 'Dias']
dias_por_mes['Mês'] = dias_por_mes['Mês'].dt.strftime('%b %Y')
print(dias_por_mes)
```

```python
fig = px.bar(
    dias_por_mes,
    x='Mês',
    y='Dias',
    title="Total de Dias Corridos por Mês",
    labels={
        "Mês": "Mês",
        "Dias": "Quantidade de Dias"
    },
    color_discrete_sequence=['#FC4C02']
)
fig.update_traces(textposition='outside')

fig.update_layout(
    title_x=0.5,
    xaxis_title="Mês",
    yaxis_title="Quantidade de Dias",
    plot_bgcolor='#C0C0C0'
)

fig.show()
```

```python
import plotly.express as px  # Outras análises relevantes

df_para_plotly = corridas_df.reset_index()

fig = px.scatter(
    df_para_plotly,
    x='Distance',
    y='Average speed',
    color='Distance',   # Colore os pontos pela distância
    size='Distance',    # Aumenta o tamanho do ponto para distâncias maiores
    hover_name='Activity Name',
    hover_data={'Activity date': '%d de %b, %Y'},
    labels={
        "Distance": "Distância da Corrida (km)",
        "Average speed": "Pace (minutos por km)"
    },
    title="Análise de Pace vs. Distância de Cada Corrida",
    color_continuous_scale='Rainbow'
)

fig.show()
```
